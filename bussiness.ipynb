```python
Global_Rank=driver.find_elements(by=By.XPATH, value='//*[@id="content_section"]/div[1]/table[1]/tbody/tr/td[1]')
Indian_Companies=driver.find_elements(by=By.XPATH, value='//*[@id="content_section"]/div[1]/table[1]/tbody/tr/td[2]')
Sales=driver.find_elements(by=By.XPATH, value='//*[@id="content_section"]/div[1]/table[1]/tbody/tr/td[3]')
Profits=driver.find_elements(by=By.XPATH, value='//*[@id="content_section"]/div[1]/table[1]/tbody/tr/td[4]')
Assests=driver.find_elements(by=By.XPATH, value='//*[@id="content_section"]/div[1]/table[1]/tbody/tr/td[5]')
Marketvalue=driver.find_elements(by=By.XPATH, value='//*[@id="content_section"]/div[1]/table[1]/tbody/tr/td[4]')
```

```python
IndianCompanies=[]
for i in range(1,len(Global_Rank)):
    td={
        'Comapany':Indian_Companies[i].text,
        'rank':Global_Rank[i].text[1:],
        'Sales':Sales[i].text,
        'Profit':Profits[i].text,
        'Assest':Assests[i].text,
        'MarketValue':Marketvalue[i].text
    }
    IndianCompanies.append(td)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(IndianCompanies)
```
