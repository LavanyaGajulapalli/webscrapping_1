```python
s=Service("D:\chromedrive\chromedriver.exe")
driver=webdriver.Chrome(service=s)
```

```python
import pandas as pd
```

```python
driver.get('https://www.zaubacorp.com/')
driver.maximize_window()
```

```python
temp = []
for i in data['CORPORATE_IDENTIFICATION_NUMBER']:
    inputElement = driver.find_element_by_id('searchid')
    inputElement.send_keys(i)#U85110TG1997NPL027669
    inputElement.send_keys(Keys.ENTER)
    ele=driver.find_elements(By.XPATH,value='//*[@id="block-system-main"]/div[2]/div[1]/div[1]/div[1]/table/tbody/tr/td[2]')
    ind=driver.find_elements(By.XPATH,value='//*[@id="block-system-main"]/div[2]/div[1]/div[1]/div[1]/table/tbody/tr/td[1]')
    td={}
    td['CIN']=i
    for j in range(len(ele)):
        td[ind[j].text]=ele[j].text
    temp.append(td)
```

```python
df=pd.DataFrame(temp)
df
```
